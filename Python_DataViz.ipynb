#**Numpy**

###Creating Arrays

```python
import numpy as np

a=np.array([1,2,3,4,5])
b=np.arange(1,6)#Creates [1,2,3,4,5]
c=np.zeros((2,3))#2x3 matrix of zeros
d=np.ones((3,2))#3x2 matrix of ones

print("Array a:",a)
print("Array b:",b)
print("2x3 Zeros:",c)
print("3x2 Ones:",d)
```

##Indexing & Slicing

```python
print("First Element of a:",a[0])
print("Last Two Elements of a:",a[-2:])
```

##Basic Arthimetic Operations

```python
print("a + 2",a + 2)
print("a * 3",a * 3)
```

###Basic Statistics

```python
print("Square of a:",np.square(a))
print("Mean of a",np.mean(a))
print("Standard Deviation",np.std(a))
```

###Reshaping Arrays

```python
matrix=np.arange(1,10).reshape((3,3))
print("Reshaping 3x3 matrix:")
matrix
```

###Transpose

```python
print("Tanspose:")
matrix.T
```

###Boolean Indexing

```python
print("Elements > 5:",matrix[matrix>5])
```

#**Pandas**

###Create a DataFrame Manually

```python
import pandas as pd

data={
    'Name':['Alice','Bob','Charlie','david','Eva'],
    'Age':[25,30,35,40,22],
    'Department':['Sales','Engineering','HR','Engineering','Sales'],
    'Salary':[50000,80000,40000,90000,45000]
}
df=pd.DataFrame(data)
df
```

###Basic Operations

```python
print("First 3 rows:")
df.head(3)
```

```python
print("Column Names:")
df.columns
```

```python
print("Data Type")
df.dtypes
```

```python
print("Describe numerical columns:")
df.describe()
```

##Filteing

```python
print("people With Salary > 50000:")
df[df['Salary'] > 50000]
```

##Sorting

```python
print("Sorting  By Age:")
df.sort_values(by='Age')
```

###Group and Aggregation

```python
grouped = df.groupby('Department')['Salary'].mean()
print("Average Salary By Department:")
grouped
```

###Adding New Columns

```python
df['Salary in Lakhs']= df['Salary']/100000
df
```

#**Matplotlib**

###Line Plot

```python
import matplotlib.pyplot as plt

years = [2019,2020,2021,2022]
revenue = [100,120,130,160]

plt.plot(years,revenue,marker='o',linestyle='--',color='purple')
plt.title("Company Revenues Over years")
plt.xlabel("Year")
plt.ylabel("Revenue {in crores}")
plt.xticks(years)
plt.grid(True)
plt.show()
```

###Bar Chart

```python
departments=df['Department'].value_counts()
departments.plot(kind='bar',color='pink')
plt.title("Employee Count By Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

###Histogram

```python
ages = df['Age']
plt.hist(ages,bins=5,color='teal',edgecolor='chocolate')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

###Pie Chart

```python
df.groupby('Department').size().plot(kind='pie',autopct='%1.1f%%')
plt.title("department Proporation")
plt.ylabel("")
plt.show()
```

#**SeaBorn:Statictical Data Visualization**

###Load Sample Dataset

```python
import seaborn as sns

tips=sns.load_dataset("tips")
tips.head()
```

```python
print("\n Info about the tips dataset:")
tips.info()

print("\n Description of the tips dataset:")
tips.describe(include='all')
```

#Pair Plot

```python
sns.pairplot(tips,hue="sex")
plt.show()
```

#Heatmap of Correlation

```python
corr=tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True,cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

#Countplot

```python
sns.countplot(x='day',data=tips,palette='Set2',hue='day',legend=False)
plt.title("Count Of Records By day")
plt.show()
```

#Boxplot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Boxplot Of Total Bill By day")
plt.show()
```
